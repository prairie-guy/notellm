### Bioinformatics: Genome Analysis

```python
%cc Locate the ncbi reference fasta file for E. Coli and save to a file. Then load the first 10 lines of the file
```

```python
from Bio import Entrez
import gzip

Entrez.email = "your.email@example.com"

accession = "U00096.3"
handle = Entrez.efetch(db="nucleotide", id=accession, rettype="fasta", retmode="text")
fasta_data = handle.read()
handle.close()

output_file = "ecoli_k12_mg1655.fasta"
with open(output_file, 'w') as f:
    f.write(fasta_data)

with open(output_file, 'r') as f:
    first_10_lines = [f.readline() for _ in range(10)]

print(''.join(first_10_lines))
```

```python
%cc calculate the CG(%)
```

```python
with open('ecoli_k12_mg1655.fasta', 'r') as f:
    lines = f.readlines()

sequence = ''.join(line.strip() for line in lines[1:])

c_count = sequence.count('C')
g_count = sequence.count('G')
total_bases = len(sequence)

gc_content = ((c_count + g_count) / total_bases) * 100

print(f"GC Content: {gc_content:.2f}%\nTotal bases: {total_bases:,}\nC count: {c_count:,}\nG count: {g_count:,}\nC+G count: {c_count + g_count:,}")
```
