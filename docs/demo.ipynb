```python
from altair.datasets import data
import polars as pl

# Import penguin dataset and convert to Polars
penguins_df = pl.from_pandas(data.penguins())
penguins_df
```

```python
%cc Create violin plots with seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.utils.capture import capture_output

# Convert to pandas for seaborn (seaborn works best with pandas)
penguins_pd = penguins_df.to_pandas()

with capture_output() as _claude_captured_output:
    # Create a figure with multiple violin plots
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('Penguin Dataset - Violin Plots by Species', fontsize=16, fontweight='bold')
    
    # Beak Length
    sns.violinplot(data=penguins_pd, x='Species', y='Beak Length (mm)', 
                   hue='Species', ax=axes[0, 0], palette='Set2', legend=False)
    axes[0, 0].set_title('Beak Length by Species')
    
    # Beak Depth
    sns.violinplot(data=penguins_pd, x='Species', y='Beak Depth (mm)', 
                   hue='Species', ax=axes[0, 1], palette='Set2', legend=False)
    axes[0, 1].set_title('Beak Depth by Species')
    
    # Flipper Length
    sns.violinplot(data=penguins_pd, x='Species', y='Flipper Length (mm)', 
                   hue='Species', ax=axes[1, 0], palette='Set2', legend=False)
    axes[1, 0].set_title('Flipper Length by Species')
    
    # Body Mass
    sns.violinplot(data=penguins_pd, x='Species', y='Body Mass (g)', 
                   hue='Species', ax=axes[1, 1], palette='Set2', legend=False)
    axes[1, 1].set_title('Body Mass by Species')
    
    plt.tight_layout()
    plt.show()

for output in _claude_captured_output.outputs:
    display(output)

# Return a summary
"Violin plots created for all numerical features grouped by penguin species"
```

```python
%cc Locate the ncbi reference fasta file for E. Coli and save to a file. Then load the first 10 lines of the file
```

```python
import urllib.request
import gzip

# Download E. coli K-12 MG1655 reference genome from NCBI
url = "https://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/005/845/GCF_000005845.2_ASM584v2/GCF_000005845.2_ASM584v2_genomic.fna.gz"

# Define output file paths
output_file_gz = "ecoli_k12_mg1655.fna.gz"
output_file = "ecoli_k12_mg1655.fna"

# Download the file
print(f"Downloading E. coli K-12 MG1655 reference genome from NCBI...")
urllib.request.urlretrieve(url, output_file_gz)
print(f"‚úì Downloaded to {output_file_gz}")

# Decompress and save
print(f"Decompressing...")
with gzip.open(output_file_gz, 'rb') as f_in:
    with open(output_file, 'wb') as f_out:
        f_out.write(f_in.read())
print(f"‚úì Saved decompressed file to {output_file}")

# Read first 10 lines
print(f"\nFirst 10 lines of {output_file}:")
print("-" * 80)
with open(output_file, 'r') as f:
    first_10_lines = [f.readline().rstrip() for _ in range(10)]
    for i, line in enumerate(first_10_lines, 1):
        print(f"{i:2d}: {line}")

first_10_lines
```

```python
%cc calculate the CG(%)
```

```python
from Bio import SeqIO

# Read the E. coli FASTA file
fasta_file = "ecoli_k12_mg1655.fna"

# Parse the FASTA file and calculate GC content
for record in SeqIO.parse(fasta_file, "fasta"):
    sequence = str(record.seq)
    
    # Count bases
    total_length = len(sequence)
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    a_count = sequence.count('A')
    t_count = sequence.count('T')
    
    # Calculate GC content
    gc_count = g_count + c_count
    gc_percentage = (gc_count / total_length) * 100
    
    # Display results
    results = {
        'Sequence ID': record.id,
        'Description': record.description,
        'Total Length (bp)': f"{total_length:,}",
        'A count': f"{a_count:,}",
        'T count': f"{t_count:,}",
        'G count': f"{g_count:,}",
        'C count': f"{c_count:,}",
        'GC count': f"{gc_count:,}",
        'GC Content (%)': f"{gc_percentage:.2f}%"
    }
    
    print("E. coli K-12 MG1655 Genome Statistics")
    print("=" * 60)
    for key, value in results.items():
        print(f"{key:20s}: {value}")
    
    # Return the GC percentage
    gc_percentage
```
