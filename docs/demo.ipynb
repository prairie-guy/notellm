```python
# Import penguin dataset using Altair 6.0+ syntax
from altair.datasets import data

# Load the penguins dataset
penguins = data.penguins()
penguins.head()
```

```python
%cc I prefer polars
```

```python
# Convert the penguins dataset to Polars
penguins_pl = pl.from_pandas(penguins)
penguins_pl
```

```python
%cc Use an Altair to show these as Violin Plots
```

```python
# Create violin plots for penguin measurements by species
from IPython.utils.capture import capture_output

with capture_output() as _claude_captured_output:
    # Body Mass violin plot
    chart1 = alt.Chart(penguins_pl).transform_density(
        'Body Mass (g)',
        as_=['Body Mass (g)', 'density'],
        groupby=['Species']
    ).mark_area(orient='horizontal', opacity=0.5).encode(
        y='Body Mass (g):Q',
        color='Species:N',
        x=alt.X('density:Q', stack='center', impute=None, title=None, axis=alt.Axis(labels=False, values=[0], grid=False, ticks=True)),
        column=alt.Column('Species:N', header=alt.Header(titleOrient='bottom', labelOrient='bottom'))
    ).properties(
        width=100,
        title='Body Mass by Species'
    )
    
    # Flipper Length violin plot
    chart2 = alt.Chart(penguins_pl).transform_density(
        'Flipper Length (mm)',
        as_=['Flipper Length (mm)', 'density'],
        groupby=['Species']
    ).mark_area(orient='horizontal', opacity=0.5).encode(
        y='Flipper Length (mm):Q',
        color='Species:N',
        x=alt.X('density:Q', stack='center', impute=None, title=None, axis=alt.Axis(labels=False, values=[0], grid=False, ticks=True)),
        column=alt.Column('Species:N', header=alt.Header(titleOrient='bottom', labelOrient='bottom'))
    ).properties(
        width=100,
        title='Flipper Length by Species'
    )
    
    # Combine charts vertically
    combined = alt.vconcat(chart1, chart2)
    display(combined)

for output in _claude_captured_output.outputs:
    display(output)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.utils.capture import capture_output

with capture_output() as _claude_captured_output:
    # Set the style
    sns.set_theme(style="whitegrid")
    
    # Create a figure with two subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Body Mass violin plot
    sns.violinplot(data=penguins_pl, x='Species', y='Body Mass (g)', 
                   hue='Species', ax=axes[0], legend=False)
    axes[0].set_title('Body Mass by Species')
    
    # Flipper Length violin plot
    sns.violinplot(data=penguins_pl, x='Species', y='Flipper Length (mm)', 
                   hue='Species', ax=axes[1], legend=False)
    axes[1].set_title('Flipper Length by Species')
    
    plt.tight_layout()
    plt.show()

for output in _claude_captured_output.outputs:
    display(output)
```
